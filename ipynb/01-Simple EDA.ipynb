```python
def distbycat(categorical,numeric):
    for cat in housing_df[categorical].unique():
        sns.distplot(housing_df[housing_df[categorical] == cat][numeric],label=cat)
    sns.plt.legend()
```

```python
def distbycatnolabel(categorical,numeric):
    for cat in housing_df[categorical].unique():
        sns.distplot(housing_df[housing_df[categorical] == cat][numeric])
```
